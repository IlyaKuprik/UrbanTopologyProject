```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np

# Создаем простой граф
G = nx.karate_club_graph()

# Расчет различных метрик центральности
degree_centrality = nx.degree_centrality(G)
eigenvector_centrality = nx.eigenvector_centrality(G, max_iter=1000)
closeness_centrality = nx.closeness_centrality(G)
betweenness_centrality = nx.betweenness_centrality(G)

# Визуализация
def visualize_graph(G, metric_values, title):
    node_color = [metric_values[node] for node in G.nodes()]
    node_size = [v * 1000 for v in node_color]
    pos = nx.spring_layout(G)
    
    plt.figure(figsize=(10, 10))
    nx.draw_networkx_nodes(G, pos, node_size=node_size, node_color=node_color, cmap=plt.cm.viridis)
    nx.draw_networkx_edges(G, pos, alpha=0.3)
    nx.draw_networkx_labels(G, pos)
    
    plt.title(title)
    plt.colorbar(plt.cm.ScalarMappable(cmap=plt.cm.viridis), label='Centrality value')
    plt.axis('off')
    plt.show()

# Визуализируем каждую метрику
visualize_graph(G, degree_centrality, "Degree Centrality")
visualize_graph(G, eigenvector_centrality, "Eigenvector Centrality")
visualize_graph(G, closeness_centrality, "Closeness Centrality")
visualize_graph(G, betweenness_centrality, "Betweenness Centrality")
```

```python
import networkx as nx

# Создание графа
G = nx.Graph()

# Добавление узлов и ребер (например, граф с четырьмя узлами и несколькими ребрами)
G.add_nodes_from([1, 2, 3, 4])
G.add_edges_from([(1, 2), (2, 3), (3, 4), (1, 3)])

# Расчет промежуточной центральности
betweenness = nx.betweenness_centrality(G)

# Вывод результатов
for node, centrality in betweenness.items():
    print(f"Node {node}: Betweenness Centrality = {centrality}")
```
